```python
# Define a dictionary containing employee data
data1 = {'Name1':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
 
# Convert the dictionary into DataFrame 
df1 = pd.DataFrame(data1)

data2 = {'Name2':['Jai', 'Prince', 'Garv', 'Anu'],
        'Age':[27, 23, 27, 30],
        'study_loc':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
 
# Convert the dictionary into DataFrame 
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df=[df1,df2]

blank_df=pd.DataFrame()

for data in df:
    data.rename(columns={data.columns[0]:'particular name'},inplace=True)
    if blank_df.empty:
        blank_df=data
    else:
        blank_df=pd.merge(blank_df,data,left_on=blank_df.columns[0],right_on=data.columns[0],how='outer')
    
```

```python
blank_df
```

```python
drop_col=['Age','Qualification']

for col in drop_col:
    merge_col=[f'{col}_x',f'{col}_y']
    blank_df[col]=blank_df[merge_col].apply(lambda row: row[0] if pd.notna(row[0]) else row[1],axis=1)
blank_df=blank_df.drop(columns=[f'{col}_x' for col in drop_col]+[f'{col}_y' for col in drop_col])
    
```

```python
blank_df
```
